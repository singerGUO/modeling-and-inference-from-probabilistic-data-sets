```python
def discrtization(l,u,x):
    length=(u-l)/100
    id=(x-l)//length
    x=l+(id+0.5)*length
    return x
    
```

```python
#A is in the report

#random generator
def rand():
    a=np.random.randint(2) 
    if a==1 :
        return 1
    else:
        return -1
```

```python
#B
X=[]
Y=[]

for i in range(100000):
    random=rand()
    X.append(random)
    s = np.random.normal(0, 1)
    Y.append(random+s)
    

mean=np.mean(X)
variance=np.var(X)
print("mean:",mean)
#this is almost the mean=0
print("variance",variance)
#very close to the 1
```

```python
#C and D
average=np.mean(Y)
std=np.std(Y)
lb=round(average-5*std,1)
ub=round(average+5*std,1)

pmf=[]
for n in Y:
    result=round(discrtization(lb,ub,n),3)
    pmf.append(result)

step=(ub-lb)/100



x = np.arange(lb+step/2,ub,step) 
Dict = {round(key,3):0 for key in x}
# Dict[0]=0
# print(Dict.keys())

for n in pmf:
    Dict[n] += 1

y = [Dict[round(key,3)]/100000 for key in x]


plt.xlim([lb,ub])
#plt.ylim([0,.015])
plt.plot(x,y)

```

```python
#E
count=0
signal=[]

for item in Y:
    if item>0:
        signal.append(1)
    else:
        signal.append(-1)
for a, b in zip(signal,Z):
    if a!=b:
        count+=1
print(count)

#probability of error
P=count/len(X)
print("the probability of it is almost percent",P)

```
