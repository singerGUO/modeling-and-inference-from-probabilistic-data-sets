```python
#A write a function that return array n by 1 arrays containing Acos(2Ï€ft) for all T
def StudentCosine(T,f,A):
    X=A*np.cos(2*T*f*np.pi )
    return X
```

```python
#B length of 10000 samples 
T=np.arange(0,1,.0001)
#generate 20 cycles 
f=20
A=1
x1=StudentCosine(T,f,A)
print(T)
print (x1)
```

```python
#A)Define two non-trivial 3-by-3 matrices. 
A = np.mat([[0, 1, 2], [3, 4, 5],[1,5,4]])
B = np.mat([[1, 0, 2],[1,1,1],[2,0,1]])
#B) Calculate  the  determinant  ofAandB
print(np.linalg.det(A))
print( np.linalg.det(B))
#C) 
#summation
print(A+B)
#multiplication
print(A*B)
#D inverse of A and B
print(np.linalg.inv(A))
print(np.linalg.inv(B))
#PartE
V, Q = LA.eig(A)
print('eigenValues :',V)
print('eigenvectors:')
print(Q)
diagonal=np.mat([[V[0],0,0],[0,V[1],0],[0,0,V[2]]])


result=Q*diagonal*np.linalg.inv(Q)
print(result)
# result is the same as matrix A except there are some deviations from python

#part f

F=np.mat([[5, 6, 7], [6, 3, 2],[7,2,1]])
F_V, F_Q = LA.eig(F)
diagonal_F=np.mat([[F_V[0],0,0],[0,F_V[1],0],[0,0,F_V[2]]])


transpose_Q= numpy.transpose(F_Q)
#QTQ=QQT=I
print(transpose_Q*F_Q)
print(F_Q*transpose_Q)


result_F_inverse=F_Q*diagonal_F*np.linalg.inv(F_Q)
result_F_transpose=F_Q*diagonal_F*transpose_Q

print(result_F_inverse)
print(result_F_transpose)
#QÎ›Qâˆ’1andQÎ›QT are the same with the original matrix.



```

2.1  Fair Coin Toss

```python
def FairCoinToss():
    if random.uniform(0,1)<0.5: return "H"
    else: return "T"
```

```python
#A toss it 10000 times because we need the value 
thislist=[]
for i in range (10000):
    thislist.append(FairCoinToss())
#B     
k=10
n=0
dictionary={}
while k<=10000:
    
    
    for i in range(k):
        if thislist[i]=="H":
            n+=1
    #store every H(k)/k as value and K as key
    dictionary[k]=n/k
    k+=10
    n=0



lists = sorted(dictionary.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()

#C

thislist=[]
for i in range (10000):
    thislist.append(FairCoinToss())
    
k=10
n=0
dictionary={}
while k<=10000:
    
    #store every H(k)/k as value and K as key
    for i in range(k):
        if thislist[i]=="T":
            n+=1
    dictionary[k]=n/k
    k+=10
    n=0



lists = sorted(dictionary.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)

plt.show()
# in this situation the limit is the probability of the tail

# As shown in the graphs they all converges to the 0.5 which the probability of the head and tail are 0.5 respectively
```

As shown in the graphs they all converges to the 0.5 which the probability of the head and tail are 0.5 respectively

2.3  Eight-sided Die Rolling

```python
#A
Dice={
    "HHH":8,
    "HHT":1,
    "HTT":2,
    "HTH":3,
    "THH":4,
    "THT":5,
    "TTT":6,
    "TTH":7
    
}
def EightSidedDieRolling():
    result=""
    for i in range(3):
        
        result+=FairCoinToss()
    return Dice[result]

print(EightSidedDieRolling())
    
```

```python
#B
thislist=[]
for i in range (10000):
    thislist.append(EightSidedDieRolling())
    
k=10
n=0
dictionary={}
while k<=10000:
    
    
    for i in range(k):
        if thislist[i]==1:
            n+=1
    dictionary[k]=n/k
    k+=10
    n=0



lists = sorted(dictionary.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples


plt.plot(x, y)
plt.show()
print(y[-1])
# it converges to 0.125 which is equivalent to  1/8
```
